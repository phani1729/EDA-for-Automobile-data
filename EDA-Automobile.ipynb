```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',50)
automobile=pd.read_csv(your_local_path+'Automobile_data.txt',na_values=["?"])
automobile.head()
```

```python
auto=automobile['city-mpg'].sort_values(ascending=False)
auto.head()
```

```python
# Top 3 makers in terms of mileage
automobile.iloc[[30,18,90]]
```

```python
automobile['normalized-losses']=automobile['normalized-losses'].fillna((automobile['normalized-losses'].mean()))
automobile.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
automobile.describe()
```

```python
# Plot showing relation between city-mpg and price for different automobiles

sns.lmplot('city-mpg', 'price', data=automobile, fit_reg=False, hue="make",aspect=2.5)
```

```python
# Boxplots showing price distribution for different car makes

plt.figure(figsize=(17,7))
sns.boxplot(x='make',y='price',data=automobile,linewidth=1)
plt.show()
```

```python
#Sorting automobile prices in descending order
price=automobile.price.sort_values(ascending=False)
price.head()
```

```python
# Bottom 5 rows in terms of price
price=automobile.price.sort_values(ascending=False)
price.tail()
```

```python
automobile.head()
```

```python
plt.figure(figsize=(18,5))
plt.subplots_adjust(hspace=0.2,wspace=0.5)
plt.subplot(1,4,1)
plt.title('city-mpg')
plt.plot(automobile['city-mpg'],'r')
plt.subplot(1,4,2)
plt.title('horsepower')
plt.plot(automobile.horsepower,'b')
plt.subplot(1,4,3)
plt.title('symboling')
plt.plot(automobile.symboling,'g')
plt.subplot(1,4,4)
plt.title('price')
plt.plot(automobile.price,'y')
plt.show()
```

```python
# Finding mileage and price
mileage=automobile[['city-mpg','price']].sort_values('price',ascending=False)
mileage.head()
```

```python
# Which manufacturer's cars is the safest based on risk rating

most_safe_car=automobile.groupby('make').agg({'symboling':[np.size,np.mean]})
most_safe_car.sort_values([('symboling','mean')],ascending=False)
```

```python
# Most preferred car based on 'fuel-type',number of doors and body style
most_prefer=pd.pivot_table(automobile,values='price',index=['num-of-doors'],columns=['fuel-type','body-style'])
most_prefer
```

```python
# Finding total count of cars based on 'num of doors,fuel-type,body-style'
count=automobile.groupby(['num-of-doors','fuel-type','body-style']).size().sort_values(ascending=False)
count
```

```python
# Which car is most priced on an average

avg_price=automobile.groupby('make').agg({'price':[np.size,np.mean]})
avg_price.sort_values([('price','mean')],ascending=False)
```

```python
# Car with highest average price and having at least '6' models

at_least=avg_price['price']['size']>=6
avg_price[at_least].sort_values([('price','mean')],ascending=False)
```

```python
#. No cars having 2,3,8,12 cylinders for diesel cars, whereas gas cars comes with every range of cylinders

cylinders=pd.pivot_table(automobile,values='city-mpg',index=['fuel-type'],columns=['num-of-cylinders'])
cylinders
```

```python
# Analysis between make, drive wheels and curb weight
# Cars with more curb weight have 'rear wheel drive' mostly(exception being AUDI which choose to have 4wd instead of rwd
# for higher curb weight cars)

bivariate_ana=pd.pivot_table(automobile,values='curb-weight',index=['make'],columns=['drive-wheels'])
bivariate_ana.sort_values('fwd',ascending=False)
```

```python
# Analysis between make, drive wheels and horsepower
trivariate_ana=pd.pivot_table(automobile,values='horsepower',index=['make'],columns=['drive-wheels'])
trivariate_ana.sort_values('fwd',ascending=False)
```

```python
# Corelation matrix between all the variables
corr=automobile.corr()
plt.figure(figsize=(10,7))
sns.heatmap(corr,annot=True)
```

```python
auto_grpby=automobile.groupby('make').mean()
auto_grpby
```

```python
# Showing relation between engine-size and price
plt.figure(figsize=(20,15))
sns.boxplot(x='engine-size',y='price',data=automobile,hue='make',linewidth=2)
plt.show()
```

```python
# Sorting values in decreasing order of engine-size

auto_engsize=automobile.groupby('make').agg({'engine-size':[np.size,np.mean],'price':[np.size,np.mean]})
auto_engsize.sort_values([('engine-size','mean')],ascending=False)
```

```python
#Sorting values in decreasing order of horsepower

auto_horse=automobile.groupby('make').agg({'horsepower':[np.size,np.mean],'price':[np.size,np.mean]})
auto_horse.sort_values([('horsepower','mean')],ascending=False)
```

```python
auto=automobile[['make','city-mpg','price']]
```

```python
# Sorting in decreasing order of city-mileage

autom = auto.pivot_table(index=['make'],aggfunc='mean').sort_values(['city-mpg','price'],ascending=False)
autom=autom.reset_index()
autom
```

```python
# Only 'Porsche' having engine at rear end, rest all having engine at front
engsize=pd.pivot_table(automobile,values='price',index=['make'],columns=['engine-location'])
engsize
```

#### Conclusion

```python
conclusion= sns.PairGrid(autom,x_vars=["city-mpg", "price"],y_vars=["make"],size=4)
conclusion.map(sns.violinplot, palette="pastel");
```
